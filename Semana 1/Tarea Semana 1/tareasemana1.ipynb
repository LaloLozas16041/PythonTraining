### Ejercicio 12
#### Escribe un programa que tome una cadena de texto y que la divida en palabras, eliminando los espacios en blanco.


```python
print ("Introduce una frase")
frase = input()
frase = frase.strip()
frase = frase.split(" ")
a = 0
b = len(frase)
while a < b:
    if frase[a] == " ":
        frase.pop(a)
        b = b-1
    elif frase[a] == "":
        frase.pop(a)
        b = b-1
    else:
        a = a+1
print (frase)
```

### Ejercicio 20
#### Escribe un programa que tome una cadena de texto y que devuelva una nueva cadena con todas las letras en mayúsculas excepto la primera letra de cada palabra, que debe ser en minúsculas.


```python
print ("Escribe una frase")
frase = input()
frase = frase.strip()
frase = frase.split(" ")
a = 0
b = len(frase)
while a < b:
    if frase[a] == " ":
        frase.pop(a)
        b = b-1
    elif frase[a] == "":
        frase.pop(a)
        b = b-1
    else:
        a = a+1
a = len(frase)
b = 0
lista1 = []
mayus = ""
while b < a:
    mayus = frase[b]
    mayus = mayus.upper()
    lista1.insert(b, mayus)
    b = b+1
a = len(lista1)
b = 0
cadena = ""
while b < a:
    palabra = lista1[b]
    c = len(lista1[b])
    minu = palabra[0]
    minu = minu.lower()
    palabra = palabra[1:c]
    cadena = cadena + minu + palabra +" "
    b = b+1
print (cadena)
```
