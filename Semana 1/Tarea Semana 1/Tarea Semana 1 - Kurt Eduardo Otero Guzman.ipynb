## Ejercicio 8

8. Escribe un programa que determine si una cadena de texto es un palíndromo o no.

```python
cadena = input("Ingresa una cadena de texto: ")

# Se eliminan los espacios y se convierte la cadena a minúsculas
cadena = cadena.replace(" ", "").lower()

# Se compara la cadena con su inversa
if cadena == cadena[::-1]:
    print("La cadena es un palíndromo")
else:
    print("La cadena no es un palíndromo")
```

### Explicacion

## Eejercicio 9

9. Crea un programa que tome una lista de números y elimine todos los duplicados.

```python
numeros = input("Dame una lista de números separados por comas: ")

# Se convierte la cadena de números en una lista de números
numeros_lista = numeros.split(",")
numeros_lista = [int(n) for n in numeros_lista]

# Se eliminan los duplicados de la lista
numeros_lista = list(set(numeros_lista))

# Se imprime la lista sin duplicados
print("Sin duplicados:", numeros_lista)
```

## Ejercicio 13

13. Crea un programa que tome dos listas de números y que devuelva una tercera lista con los elementos comunes a ambas.

```python
lista1 = input("Ingresa una lista de números separados por comas: ")
lista2 = input("Ingresa otra lista de números separados por comas: ")

# Se convierten las cadenas en listas de números
lista1 = lista1.split(",")
lista1 = [int(n) for n in lista1]

lista2 = lista2.split(",")
lista2 = [int(n) for n in lista2]

# Se crea una lista vacía para almacenar los elementos comunes
comunes = []

# Se recorren los elementos de la lista1 y se comparan con los de la lista2
for elemento in lista1:
    if elemento in lista2 and elemento not in comunes:
        comunes.append(elemento)

# Se imprime la lista de elementos comunes
print("Los elementos comunes son:", comunes)
```

## Explicacion

pedimos al usuario que nos de dos listas separadas por comas, separamos la cadena en numeros, creamos una lista para almacenar los comunes, recorremos la lista1 y los comparamos con la lista2, imprimimos

18. Escribe un programa que tome una lista de palabras y que devuelva una lista con todas las palabras en mayúsculas.

```python
palabras = input("Ingresa una lista de palabras separadas por comas: ")

# Se convierte la cadena de palabras en una lista de palabras
palabras_lista = palabras.split(",")

# Se crea una lista vacía para almacenar las palabras en mayúsculas
mayusculas = []

# Se recorren las palabras en la lista y se agregan a la lista de mayúsculas
for palabra in palabras_lista:
    mayusculas.append(palabra.upper())

# Se imprime la lista de palabras en mayúsculas
print("La lista de palabras en mayúsculas es:", mayusculas)

```

## Ejercicio 20

20. Escribe un programa que tome una cadena de texto y que devuelva una nueva cadena con todas las letras en mayúsculas excepto la primera letra de cada palabra, que debe ser en minúsculas.

```python
cadena = input("Ingresa una cadena de texto: ")

# Se divide la cadena en palabras y se convierte la primera letra de cada palabra en minúscula
palabras = cadena.split()
palabras = [palabra[0].lower() + palabra[1:] for palabra in palabras]

# Se convierte toda la cadena en mayúsculas
cadena_mayusculas = cadena.upper()

# Se reemplazan las primeras letras de cada palabra por las letras en minúscula
for i in range(len(palabras)):
    buscar = palabras[i].upper()
    reemplazar = palabras[i]
    cadena_mayusculas = cadena_mayusculas.replace(buscar, reemplazar, 1)

# Se imprime la cadena con todas las letras en mayúsculas excepto la primera letra de cada palabra
print("La cadena en mayúsculas con excepción de la primera letra de cada palabra es:", cadena_mayusculas)
```
