```python
#Ejercicio 9
#Crea un programa que tome una lista de números y elimine todos los duplicados.
```

```python
lista=[1,1,2,2,3,4,4,5]
nueva=[]
for n in lista:
    if n not in nueva:
       nueva.append(n)
print(nueva)
```
