```python
#4
numero  = input("ingresa un numero: ")
es_Primo = True
for x in range(2,int(numero)+1):
    if int(numero) % int(x) == 0:    
        if not int(numero) == x:
            es_Primo = False
    
if es_Primo:
    print (str(numero)  + " es numero primo")
else:
    print (str(numero)  + " no es numero primo")
```
