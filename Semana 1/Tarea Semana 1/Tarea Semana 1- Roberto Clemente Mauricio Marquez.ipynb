### Ejercicio 19

Crea un programa que tome una lista de números y que devuelva la suma de aquellos que son mayores a un valor específico ingresado por el usuario.

```python
numeros = [5, 6, 8, 12, 30,1,3]
print(numeros)
valor = input("Dame el numero desde donde iniciara la suma: ")
valor = int(valor)
suma = 0
bandera = 0

for i in numeros:
    if valor == i or bandera == 1:
        bandera = 1
        suma = suma + i
print("La suma despues del numero", valor, "es: ", suma)
```

#### Explicación
Primero inicializamos la suma y la bandera para marcar el inicio de la suma, entramos al ciclo y validamos si el numero que se dio es igual al de la lista o si la bandera esta encendida, se realizara la suma apartir de ahi, hasta terminar el ciclo y terminar la suma total

### Ejercicio 20

Escribe un programa que tome una cadena de texto y que devuelva una nueva cadena con todas las letras en mayúsculas excepto la primera letra de cada palabra, que debe ser en minúsculas.

```python
palabras = input("Cadena de Texto a convertir: ")
print(palabras)
palabras = palabras.split()
lista = []
for i in palabras:
    mayuscula = i.upper()
    minuscula = i.lower()
    minuscula1 = minuscula[0]
    
    cadena = list(mayuscula)
    cadena[0] = minuscula1
    nueva = "".join(cadena) 
    lista.append(nueva)
    final = " ".join(lista)
    
print(final)
```

#### Explicación
Convertimos es string en una lista, inicializamos la lista nueva, entramos al ciclo para convertir cada palabra de la lista en mayusculas, convertimos la misma palabra en minusculas, sacamos la primer letra de la palabra en minusculas, convertimos la palabra en una lista para poder intercambiar la primer letra de las mayusculas por una minuscula, y volvemos a convertirla en string, y la integramos en una nueva lista ya convertida y finalmente esta nueva lista la convertimos en una nueva cadena y la imprmimimos.
