### El valor futuro es el valor alcanzado por un determinado capital al final del período determinado (para el ejemplo usaremos la fórmula del interés compuesto). Para calcularlo se utiliza la siguiente fórmula

##### VF=VA* ( 1+i )n

### El valor presente de una inversión es cuando calculamos el valor actual que tendrá una determinada cantidad que recibiremos o pagaremos en un futuro. Para calcularlo se utiliza la siguiente fórmula

##### VA=VF( 1+i )n 

### Donde: 
 ##### VF es el valor futuro, 
 ##### VA es el valor actual o inicial de la inversión, 
 ##### i es el tipo de interés y 
 ##### n es número de años de la inversión
 
(A) Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor futuro de la inversión cada año del periodo indicado.

(B) Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor actual del capital cada año del periodo indicado.

```python
def calcular_valor_futuro_interactivo():
    # Solicitar al usuario que ingrese el capital inicial
    while True:
        try:
            capital_inicial = float(input("Ingrese el capital inicial (número positivo): "))
            if capital_inicial >= 0:
                break
            else:
                print("Por favor, ingrese un número positivo.")
        except ValueError:
            print("Por favor, ingrese un número válido.")

    # Solicitar al usuario que ingrese el tipo de interés
    while True:
        try:
            tipo_de_interes = float(input("Ingrese el tipo de interés (porcentaje) como número decimal (ejemplo: 0.05 para 5%): "))
            if 0 <= tipo_de_interes <= 1:
                break
            else:
                print("Por favor, ingrese un número decimal entre 0 y 1.")
        except ValueError:
            print("Por favor, ingrese un número válido.")

    # Solicitar al usuario que ingrese el número de años
    while True:
        try:
            num_anos = int(input("Ingrese el número de años (número entero no negativo): "))
            if num_anos >= 0:
                break
            else:
                print("Por favor, ingrese un número entero no negativo.")
        except ValueError:
            print("Por favor, ingrese un número entero válido.")

    # Calcular y mostrar el valor futuro de la inversión para cada año del periodo indicado
    print("Año\tValor Futuro")
    for ano in range(1, num_anos + 1):
        valor_futuro = capital_inicial * (1 + tipo_de_interes) ** ano
        print(f"{ano}\t{valor_futuro:.2f}")

# Ejemplo de uso de la función
calcular_valor_futuro_interactivo()
```

### 2.- El archivo cotización.csv (archivo adjunto ) contiene las cotizaciones de empresas con las siguientes columnas: 
#### Nombre (nombre de la empresa), 
#### Final (precio de la acción al cierre de bolsa), 
#### Máximo (precio máximo de la acción durante la jornada), 
#### Mínimo (precio mínimo de la acción durante la jornada), 
#### Volumen (Volumen al cierre de bolsa), 
#### Efectivo (capitalización al cierre en miles de euros).


(A) Construir una función reciba el fichero de cotizaciones y devuelva un diccionario con los datos del fichero por columnas.

(B) Construir una función que reciba el diccionario devuelto por la función anterior y cree un fichero en formato csv con el mínimo, el máximo y la media de cada columna

```python
import csv

def leer_cotizaciones(fichero):
    # Crear un diccionario vacío para almacenar los datos por columnas
    datos = {
        "nombre": [],
        "final": [],
        "maximo": [],
        "minimo": [],
        "volumen": [],
        "efectivo": []
    }

    # Leer el fichero CSV y guardar los datos en el diccionario
    with open(fichero, 'r') as archivo:
        lector_csv = csv.reader(archivo)
        next(lector_csv)  # Saltar la fila de encabezados

        for fila in lector_csv:
            datos["nombre"].append(fila[0])
            datos["final"].append(float(fila[1]))
            datos["maximo"].append(float(fila[2]))
            datos["minimo"].append(float(fila[3]))
            datos["volumen"].append(float(fila[4]))
            datos["efectivo"].append(int(fila[5]))

    return datos

# Ejemplo de uso de la función
fichero = "cotizaciones.csv"
datos_cotizaciones = leer_cotizaciones(fichero)
print(datos_cotizaciones)

```

```python
import csv

def calcular_estadisticas_columnas(datos):
    estadisticas = {}
    for columna in datos:
        if columna != "nombre":
            minimo = min(datos[columna])
            maximo = max(datos[columna])
            media = sum(datos[columna]) / len(datos[columna])
            estadisticas[columna] = {"min": minimo, "max": maximo, "mean": media}
    return estadisticas

def guardar_estadisticas_csv(estadisticas, fichero_salida):
    with open(fichero_salida, 'w', newline='') as archivo_salida:
        campos = ['columna', 'min', 'max', 'mean']
        escritor_csv = csv.DictWriter(archivo_salida, fieldnames=campos)
        escritor_csv.writeheader()
        for columna, valores in estadisticas.items():
            fila = {"columna": columna, **valores}
            escritor_csv.writerow(fila)

# Ejemplo de uso de las funciones
fichero = "cotizaciones.tsv"
datos_cotizaciones = leer_cotizaciones(fichero)
estadisticas = calcular_estadisticas_columnas(datos_cotizaciones)
fichero_salida = "estadisticas.csv"
guardar_estadisticas_csv(estadisticas, fichero_salida)
```

### 3.- Escribir un programa para gestionar un listado telefónico con los nombres y los teléfonos de los clientes de una empresa. 
 (A) El programa debe crear el archivo TXT con la lista si no existe
 (B) El programa debe  tener una función para consultar el teléfono de un cliente e imprimirlo en pantalla.
 (C) El programa debe  tener una función para  añadir el teléfono de un nuevo cliente al archivo TXT.
 (D) El programa debe tener una función que permita visualizar todo el listado telefónico
 (E) El programa debe tener una función que guarde esta lista en un archivo CSV.


```python
import csv
import os

ARCHIVO_TXT = "listado_telefonico.txt"
ARCHIVO_CSV = "listado_telefonico.csv"

def crear_lista_telefonica():
    if not os.path.exists(ARCHIVO_TXT):
        with open(ARCHIVO_TXT, 'w') as archivo:
            archivo.write("")

def consultar_telefono(nombre):
    with open(ARCHIVO_TXT, 'r') as archivo:
        for linea in archivo:
            cliente, telefono = linea.strip().split(';')
            if cliente.lower() == nombre.lower():
                return telefono
    return None

def añadir_telefono(nombre, telefono):
    with open(ARCHIVO_TXT, 'a') as archivo:
        archivo.write(f"{nombre};{telefono}\n")

def visualizar_listado():
    with open(ARCHIVO_TXT, 'r') as archivo:
        for linea in archivo:
            cliente, telefono = linea.strip().split(';')
            print(f"{cliente}: {telefono}")

def guardar_en_csv():
    with open(ARCHIVO_TXT, 'r') as archivo_txt, open(ARCHIVO_CSV, 'w', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(['Nombre', 'Teléfono'])

        for linea in archivo_txt:
            cliente, telefono = linea.strip().split(';')
            escritor_csv.writerow([cliente, telefono])

# Ejemplo de uso de las funciones
crear_lista_telefonica()
añadir_telefono("Juan Pérez", "555-1234")
añadir_telefono("María García", "555-5678")
telefono = consultar_telefono("Juan Pérez")
if telefono:
    print(f"Teléfono de Juan Pérez: {telefono}")
else:
    print("Cliente no encontrado")
visualizar_listado()
guardar_en_csv()
```

### 4.- El archivo calificaciones.csv (archivo adjunto ) contiene las calificaciones de un curso. Durante el curso se realizaron dos exámenes parciales de teoría y un examen de prácticas. Escribir un programa que contenga las siguientes funciones:

(A) Una función que reciba el archivo de calificaciones y devuelva una lista de diccionarios, donde cada diccionario contiene la información de los exámenes y la asistencia de un alumno. La lista tiene que estar ordenada por apellidos.

(B) Una función que reciba una lista de diccionarios como la que devuelve la función anterior y añada a cada diccionario una nueva llave con la nota final del curso.  Teniendo en cuenta los siguientes criterios de evaluación 
    (A) Practicas 10 %
    (B) Practica Final 30% 
    (C) Promedio de Parciales 1 y 2  20%
    (D) Promedio de Exámenes 1 y 2 40%

(C) Una función que reciba una lista de diccionarios como la que devuelve la función anterior y devuelva dos listas, una con los alumnos aprobados y otra con los alumnos reprobados. Para aprobar el curso toma en cuenta lo siguiente : 
    (A) La asistencia tiene que ser mayor o igual que el 75%
    (B) La nota final debe ser mayor o igual a 8.0


```python
import csv

# (A)
def leer_calificaciones(archivo):
    with open(archivo, 'r', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile, delimiter='\t')
        alumnos = sorted(reader, key=lambda x: x['Apellidos'])
    return alumnos

# (B)
def calcular_nota_final(alumnos):
    for alumno in alumnos:
        practicas = float(alumno['Practicas']) * 0.1
        practica_final = float(alumno['Practica Final']) * 0.3
        parciales = (float(alumno['Parcial1']) + float(alumno['Parcial2'])) / 2 * 0.2
        examenes = (float(alumno['Examen1']) + float(alumno['Examen2'])) / 2 * 0.4
        nota_final = practicas + practica_final + parciales + examenes
        alumno['Nota Final'] = nota_final
    return alumnos

# (C)
def clasificar_alumnos(alumnos):
    aprobados = []
    reprobados = []

    for alumno in alumnos:
        asistencia = float(alumno['Asistencia'].strip('%')) / 100
        nota_final = alumno['Nota Final']
        if asistencia >= 0.75 and nota_final >= 8.0:
            aprobados.append(alumno)
        else:
            reprobados.append(alumno)

    return aprobados, reprobados


# Uso de las funciones
archivo = 'calificaciones.csv'
alumnos = leer_calificaciones(archivo)
alumnos_con_nota = calcular_nota_final(alumnos)
aprobados, reprobados = clasificar_alumnos(alumnos_con_nota)

print("Aprobados:")
for a in aprobados:
    print(a['Apellidos'], a['Nombre'], a['Nota Final'])

print("\nReprobados:")
for r in reprobados:
    print(r['Apellidos'], r['Nombre'], r['Nota Final'])
```

### -Explication

(A) leer_calificaciones(archivo):
Esta función toma como entrada el nombre del archivo de calificaciones y devuelve una lista de diccionarios con la información de los alumnos ordenada por apellidos. Utiliza el módulo csv para leer el archivo y el método DictReader para convertir cada fila en un diccionario.

(B) calcular_nota_final(alumnos):
Esta función toma como entrada la lista de diccionarios de alumnos generada por la función anterior y agrega a cada diccionario una nueva llave con la nota final del curso. Se calcula la nota final siguiendo los criterios de evaluación mencionados.

(C) clasificar_alumnos(alumnos):
Esta función toma como entrada la lista de diccionarios de alumnos con la nota final y devuelve dos listas, una con los alumnos aprobados y otra con los alumnos reprobados. Un alumno es considerado aprobado si su asistencia es mayor o igual al 75% y su nota final es mayor o igual a 8.0.

Finalmente, el programa utiliza estas funciones para leer el archivo de calificaciones, calcular las notas finales y clasificar a los alumnos en aprobados y reprobados, mostrando ambos grupos en pantalla.
