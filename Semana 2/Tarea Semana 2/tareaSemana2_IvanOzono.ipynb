Ejercicio 1

Las matemáticas financieras, resumidas en una frase, las podríamos definir como la rama de las matemáticas que estudia los flujos de dinero a través del tiempo. Básicamente se presupone que el dinero tiene menos valor en el futuro que en el presente, ya sea por un tema inflacionario o por la preferencia natural de las personas a priorizar el consumo presente.

El valor futuro es el valor alcanzado por un determinado capital al final del período determinado (para el ejemplo usaremos la fórmula del interés compuesto). Para calcularlo se utiliza la siguiente fórmula VF=VA* ( 1+i )n

El valor presente de una inversión es cuando calculamos el valor actual que tendrá una determinada cantidad que recibiremos o pagaremos en un futuro. Para calcularlo se utiliza la siguiente fórmula

VA=VF( 1+i )n 

Donde 
 VF es el valor futuro, 
 VA es el valor actual o inicial de la inversión, 
 i es el tipo de interés y 
 n es número de años de la inversión.

Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor futuro de la inversión cada año del periodo indicado.
Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor actual del capital cada año del periodo indicado.


```python

def calcular_valor_futuro(capital, tasa_interes, num_anios):

    print("Año\tValor Futuro")
    for anio in range(1, num_anios + 1):
        valor_futuro = (capital * ((1 + tasa_interes) ** anio))
        print(f"{anio}\t{valor_futuro:.2f}")

def calcular_valor_presente(capital, tasa_interes, num_anios):
    print("Año\tValor Presente")
    for anio in range(1, num_anios + 1):
        valor_presente = capital / ((1 + tasa_interes) ** anio)
        print(f"{anio}\t{valor_presente:.2f}")
   


capital = float(input("Ingrese el capital inicial: "))
tasa_interes = float(input("Ingrese la tasa de interés anual (decimal): "))
num_anios = int(input("Ingrese el número de años de la inversión: "))

calcular_valor_futuro(capital, tasa_interes, num_anios)
calcular_valor_presente(capital, tasa_interes, num_anios)

```

```python
import csv

# Crear un diccionario vacio
dict_datos = {
    "Nombre": [],
    "Final": [],
    "Máximo": [],
    "Mínimo": [],
    "Volumen": [],
    "Efectivo": []
}

# Abrir y leer el archivo 
with open("cotizacion.csv", "r") as archivo:
    # Crear el  CSV reader object
    lector_csv = csv.reader(archivo)
    # Saltarse la primera linea 
    next(lector_csv)
    # Iterar por las lineas restantes
    for valores in lector_csv:
        # Append cada valor de la lista en el diccionario
        dict_datos["Nombre"].append(valores[0])
        dict_datos["Final"].append(valores[1])
        dict_datos["Máximo"].append(valores[2])
        dict_datos["Mínimo"].append(valores[3])
        dict_datos["Volumen"].append(valores[4])
        dict_datos["Efectivo"].append(valores[5])
    
    # Funcion que remueve los puntos, y reemplaza las comas por punto
def clean_value(value):
    value = value.replace('.', '').replace(',', '.')
    return value

# Iterar a traves del diccionario
for key, values in dict_datos.items():
    # Verificar si el valor es un string
    if isinstance(values, str):
        # Limpiar el valor
        values = clean_value(values)
        # ACtualizar el valor ya modificado
        dict_datos[key] = values

    # Iterar a traves de lista de valores
    values = [clean_value(val) if val.replace('.', '').replace(',', '').isdigit() else val for val in values]
    # Anadir el separador para miles
    values = [format(float(val), ',.2f') if val.replace('.', '').replace(',', '').isdigit() else val for val in values]

    # Actualizar el diccionario con la lista modificada de valores
    dict_datos[key] = values

# Print el diccionario
print(dict_datos)

def generar_reporte_csv(dict_datos):
    # Obtener las llaves del diccionario excepto la primera ('Nombre')
    columnas = list(dict_datos.keys())[1:]
    
    # Crear una lista para almacenar los resultados
    resultados = []
    
    # Calcular los valores para cada columna
    for columna in columnas:
        # Convertir los valores de string a números (float o int) según corresponda
        valores = [float(valor.replace(',', '')) if ',' in valor else float(valor) for valor in dict_datos[columna]]
        
        # Calcular los valores requeridos
        minimo = min(valores)
        maximo = max(valores)
        media = sum(valores) / len(valores)
        
        # Agregar los resultados a la lista
        resultados.append([columna, minimo, maximo, media])
    
    # Crear y escribir en el archivo CSV
    with open('resultados.csv', 'w', newline='') as archivo_csv:
        writer = csv.writer(archivo_csv)
        writer.writerow(['Columna', 'Mínimo', 'Máximo', 'Media'])
        writer.writerows(resultados)



generar_reporte_csv(dict_datos)



```

```python
import csv
import os

archivo = "listado.txt"

# Función para verificar si el archivo existe y crearlo en caso contrario
def verificar_archivo():
    if not os.path.exists(archivo):
        with open(archivo, "w") as f:
            pass

# Función para consultar el teléfono de un cliente
def consultar_telefono(nombre):
    with open(archivo, "r") as f:
        for linea in f:
            if nombre in linea:
                nombre, telefono = linea.strip().split(", ")
                return nombre,telefono
    return None

# Función para añadir el teléfono de un nuevo cliente al archivo
def agregar_cliente(nombre, telefono):
    with open(archivo, "a") as f:
        f.write(nombre + ", " + telefono + "\n")

# Función para visualizar todo el listado telefónico
def visualizar_listado():
    with open(archivo, "r") as f:
        for linea in f:
            print(linea.strip())

# Función para guardar la lista en un archivo CSV
def guardar_csv(nombre_archivo):
    with open(nombre_archivo, "w", newline="") as f:
        escritor_csv = csv.writer(f)
        with open(archivo, "r") as f:
            for linea in f:
                nombre, telefono = linea.strip().split(", ")
                escritor_csv.writerow([nombre, telefono])

verificar_archivo()

# Añadir nuevos clientes
nombre= (input("Agregar nombre de cliente: "))
tel=(input("Agregar Telefono del cliente: "))
agregar_cliente(nombre, tel)
print("_______________________________")

# Consultar teléfonos
print("____CONSULTAR TELEFONOS_________")
print(consultar_telefono("Juan Perez"))
print(consultar_telefono("Maria Rodriguez"))
nombre_consulta=input("Nombre del cliente a consultar: ")
print(consultar_telefono(nombre_consulta))
print("_______________________________")

# Visualizar todo el listado
print("___VISUALIZAR LISTADO_____________")
visualizar_listado()
print("_______________________________")

# Guardar la lista en un archivo CSV
guardar_csv("listado.csv")
```

Ejercicio 4

El archivo calificaciones.csv (archivo adjunto ) contiene las calificaciones de un curso. 

Durante el curso se realizaron dos exámenes parciales de teoría y un examen de prácticas. Escribir un programa que contenga las siguientes funciones:

- Una función que reciba el archivo de calificaciones y devuelva una lista de diccionarios, donde cada diccionario contiene la información de los exámenes y la asistencia de un alumno.
-  La lista tiene que estar ordenada por apellidos.

- Una función que reciba una lista de diccionarios como la que devuelve la función anterior y añada a cada diccionario una nueva llave con la nota final del curso.  Teniendo en cuenta los siguientes criterios de evaluación:
   
    a - Practicas 10 %
    b - Practica Final 30% 
    c - Promedio de Parciales 1 y 2  20%
    d - Promedio de Exámenes 1 y 2 40%

- Una función que reciba una lista de diccionarios como la que devuelve la función anterior y devuelva dos listas, una con los alumnos aprobados y otra con los alumnos reprobados. Para aprobar el curso toma en cuenta lo siguiente :
   
    a - La asistencia tiene que ser mayor o igual que el 75%
    b - La nota final debe ser mayor o igual a 8.0


```python
import csv

def leer_calificaciones(archivo):
    with open(archivo, newline='') as csvfile:
        next(csvfile)  # omitir la primera fila del archivo
        reader = csv.DictReader(csvfile)
        calificaciones = [row for row in reader]
    return sorted(calificaciones, key=lambda x: x['Apellidos'])



def calcular_calificaciones_finales(calificaciones):
    for c in calificaciones:
        c['Calificacion_Final'] = (float(c['Practicas']) * 0.1 + 
                                   float(c['Practica Final']) * 0.3 + 
                                   (float(c['Parcial1']) + float(c['Parcial2'])) / 2 * 0.2 + 
                                   (float(c['Examen1']) + float(c['Examen2'])) / 2 * 0.4)
    return(calificaciones)

def clasificar_alumnos(calificaciones):
    aprobados = []
    reprobados = []
    for c in calificaciones:
        asistencia = float(c['Asistencia'][:-1])
        calificacion_final = c['Calificacion_Final']
        if asistencia >= 75 and calificacion_final >= 8.0:
            aprobados.append({'Nombre': c['Nombre'], 'Calificacion_Final': calificacion_final})
        else:
            reprobados.append({'Nombre': c['Nombre'], 'Calificacion_Final': calificacion_final})
    return aprobados, reprobados

calificaciones = leer_calificaciones('calificaciones.csv')
calificanes_finales =calcular_calificaciones_finales(calificaciones)
print(calificanes_finales)
aprobados, reprobados = clasificar_alumnos(calificaciones)

print("Alumnos aprobados:")
for a in aprobados:
    print(f"{a['Nombre']}: {a['Calificacion_Final']:.2f}")

print("\nAlumnos reprobados:")
for a in reprobados:
    print(f"{a['Nombre']}: {a['Calificacion_Final']:.2f}")

```
