### **Ejercicio 1**
#### Enunciado
Las matemáticas financieras, resumidas en una frase, las podríamos definir como la rama de las matemáticas que estudia los flujos de dinero a través del tiempo. Básicamente se presupone que el dinero tiene menos valor en el futuro que en el presente, ya sea por un tema inflacionario o por la preferencia natural de las personas a priorizar el consumo presente.

El valor futuro es el valor alcanzado por un determinado capital al final del período determinado (para el ejemplo usaremos la fórmula del interés compuesto). Para calcularlo se utiliza la siguiente fórmula
VF=VA* ( 1+i )^n

El valor presente de una inversión es cuando calculamos el valor actual que tendrá una determinada cantidad que recibiremos o pagaremos en un futuro. Para calcularlo se utiliza la siguiente fórmula

VA=VF/( 1+i )^n 

Donde 
> * VF es el valor futuro, 
* VA es el valor actual o inicial de la inversión, 
* i es el tipo de interés y 
* n es número de años de la inversión.

1. Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor futuro de la inversión cada año del periodo indicado.
2. Crear una función que reciba como entrada un capital, un tipo de interés y un número de años, y muestre por pantalla el valor actual del capital cada año del periodo indicado.

```python
def valfuturo(capital, interes, tiempo):
  conta = 1
  while conta <= tiempo:
    print(f'{conta} => {capital * (1 + interes) ** conta}')
    conta += 1
  return ''

def valactual(capital, interes, tiempo):
  conta = 1
  while conta <= tiempo:
    print(f'{conta} => {capital / (1 + interes) ** conta}')
    conta += 1
  return ''

captl = float(input('Escribe el capital: '))
intr = float(input('Escribe el interes: '))
tmp = float(input('Escribe los anios: '))

print('---- Valor Futuro ----')
valfuturo(captl, intr, tmp)

print('---- Valor Actual ----')
valactual(captl, intr, tmp)
```

5. El programa debe tener una función que guarde esta lista en un archivo CSV.

```python
import csv

#esta funcion cumple los puntos 1 y 3
def agregarnuevo(archivo,cliente,telefono):
  try:
    txtclientes = open(archivo, "r")
    leido = txtclientes.read()
    txtclientes = open(archivo, "a")
  except:
    txtclientes = open(archivo, "w")
  txtclientes.write(cliente + "," + telefono + '\n')
  txtclientes.close()

def busca(archivo,cliente):
  try:
    txtclientes = open(archivo, "r")
    lineas = txtclientes.readlines()
    #print(leido)
    for linea in lineas:
      if cliente in linea:
        print(linea)
  except:
    "El archivo no se encuentra o no puede ser leido"
  txtclientes.close()

def imprimetodo(archivo):
  try:
    txtclientes = open(archivo, "r")
    lineas = txtclientes.readlines()
    for linea in lineas:
      print(linea)
  except:
    "El archivo no se encuentra o no puede ser leido"
  txtclientes.close()

#quitar el \n
def txtacsv(archivo): 
  with open("clientes.csv",'a',newline="") as csvfile:
    escritor_csv = csv.writer(csvfile)
    try:
      txtclientes = open(archivo, "r")
      lineas = txtclientes.readlines()
      for linea in lineas:
        info = linea.split(',')
        escritor_csv.writerow(info)
    except:
      "El archivo no se encuentra o no puede ser leido"
  txtclientes.close()       

agregarnuevo("clientes.txt","Alicia Andrade","1111111111")
agregarnuevo("clientes.txt","Bruno Bose","2222222222")
agregarnuevo("clientes.txt","Claudia Colon","3333333333")
agregarnuevo("clientes.txt","Daniel Diaz","4444444444")
busca("clientes.txt","Bruno Bose")
print("------------------------")
imprimetodo("clientes.txt")
txtacsv("clientes.txt")
```

### **Ejercicio 4**

#### Enunciado
El archivo calificaciones.csv (archivo adjunto ) contiene las calificaciones de un curso. Durante el curso se realizaron dos exámenes parciales de teoría y un examen de prácticas. Escribir un programa que contenga las siguientes funciones:

1. Una función que reciba el archivo de calificaciones y devuelva una lista de diccionarios, donde cada diccionario contiene la información de los exámenes y la asistencia de un alumno. La lista tiene que estar ordenada por apellidos.
2. Una función que reciba una lista de diccionarios como la que devuelve la función anterior y añada a cada diccionario una nueva llave con la nota final del curso.  Teniendo en cuenta los siguientes criterios de evaluación: 
> 1. Practicas 10 %
> 2. Practica Final 30% 
> 3. Promedio de Parciales 1 y 2  20%
> 4. Promedio de Exámenes 1 y 2 40%
3. Una función que reciba una lista de diccionarios como la que devuelve la función anterior y devuelva dos listas, una con los alumnos aprobados y otra con los alumnos suspensos. Para aprobar el curso:
> 1. La asistencia tiene que ser mayor o igual que el 75%
> 2. La nota final debe ser mayor o igual a 8.0
 

```python
import csv

def abrir_archivo(nombre):
  milista = []
  listadic = []
  with open(nombre, newline='') as archivo:
    lector = csv.reader(archivo)
    lineas = list(lector)
    listatitulos = lineas[1]
    for linea in lineas:
      if linea != lineas[0] and linea != lineas[1]:
        milista.append(linea)
    milista.sort()
    for listita in milista:
      dicnuevo = {}
      varcontador = 0
      for titulo in listatitulos:
        dicnuevo[titulo] = listita[varcontador]
        varcontador += 1
      listadic.append(dicnuevo)
  return listadic

def notafinal(listadic):
  for dixi in listadic:
    nota = float(dixi["Practicas"]) * 0.1 + float(dixi["Practica Final"]) * 0.3 + (float(dixi["Parcial1"]) + float(dixi["Parcial2"])) / 2 * 0.2 + (float(dixi["Examen1"]) + float(dixi["Examen2"])) / 2 * 0.4
    dixi['Nota Final'] = nota

def califica(listadic):
  aprobados = []
  reprobados = []
  for alumno in listadic:
    if float(alumno['Asistencia'].replace("%","")) >= 75 and float(alumno['Nota Final']) >= 8:
      aprobados.append(alumno)
    else:
      reprobados.append(alumno)
  calificados = []
  calificados.append(aprobados)
  calificados.append(reprobados)
  return calificados

lstdiccionarios = abrir_archivo("calificaciones.csv")
#for elem in lstdiccionarios:
#  print(elem)
notafinal(lstdiccionarios)
final = califica(lstdiccionarios)
for elem in final:
  print(elem)
```
