    El programa debe tener una función que permita visualizar todo el listado telefónico
    El programa debe tener una función que guarde esta lista en un archivo CSV.



```python
def get_tel(file, cliente):
    
    try: 
        f = open(file, 'r')
    except FileNotFoundError:
        return('¡El fichero ' + file + ' no existente!\n')
    else:
        directory = f.readlines()
        f.close()
        directory = dict([tuple(line.split(',')) for line in directory])
        if cliente in directory:
            return directory[cliente]
        else:
            return('¡El cliente ' + cliente + ' no existe!\n')


def add_tel(file, cliente, telf):
    
    try: 
        f = open(file, 'a')
    except FileNotFoundError:
        return('¡El fichero ' + file + ' no existe!\n')
    else:
        f.write(cliente + ',' + telf + '\n')
        f.close()
        return('El teléfono se ha añadido.\n')

def remove_tel(file, cliente):
    
    try: 
        f = open(file, 'r')
    except FileNotFoundError:
        return('¡El fichero ' + file + ' no existe!\n')
    else:
        directory = f.readlines()
        f.close()
        directory = dict([tuple(line.split(',')) for line in directory])
        if cliente in directory:
            del directory[cliente]
            f = open(file, 'w')
            for name, telf in directory.items():
                f.write(name + ',' + telf)
            f.close()
            return ('¡El cliente se ha eliminado!\n')
        else:
            return('¡El cliente ' + cliente + ' no existe!\n')


def create_directory(file):
    
    import os
    if os.path.isfile(file):
        answer = input('El fichero ' + file + ' ya existe. ¿Desea vaciarlo (S/N)? ')
        if answer == 'N': 
            return 'No se creo el fichero porque ya existe.\n'
    f = open(file, 'w')
    f.close()
    return 'Se ha creado el fichero.\n'
     

def menu():
   
    print('Gestión del listado telefónico.\n')
   
    print('1 Consultar un cliente')
    print('2 Añadir un cliente')
    print('3 Eliminar un cliente')
    print('4 Crear el listado')
    print('0 Finalizar')
    option = input('Introduzca el número de la opción deseada: ')
    return option


def directory():
   
    file = 'listado.txt' 
    while True:
        option = menu()
        if option == '1':
            name = input('Introduce el nombre del cliente: ')
            print(get_tel(file, name))
        elif option == '2':
            name = input('Introduce el nombre del cliente: ')
            telf = input('Introduce el teléfono del cliente: ')
            print(add_tel(file, name, telf))
        elif option == '3':
            name = input('Introduce el nombre del cliente: ')
            print(remove_tel(file, name))
        elif option == '4':
            print(create_directory(file))
        else:
            break
    return

directory()
```
