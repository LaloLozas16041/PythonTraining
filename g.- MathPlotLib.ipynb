# Ejercicio Titanic

El fichero [titanic.csv] contiene información sobre los pasajeros del Titanic. Crear un dataframe con Pandas y a partir de él generar los siguientes diagramas.

1. Diagrama de sectores con los fallecidos y supervivientes.
2. Histograma con las edades.
3. Diagrama de barras con el número de personas en cada clase.
4. Diagrama de barras con el número de personas fallecidas y supervivientes en cada clase.
5. Diagrama de barras con el número de personas fallecidas y supervivientes acumuladas en cada clase.

# Solución

```python
%matplotlib inline
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_titanic = pd.read_csv('titanic.csv',index_col=0)
df_titanic
```

## 1. Diagrama de sectores con los fallecidos y supervivientes.


```python
survived=df_titanic.Survived.value_counts().to_list()
survived_labels=['Muertos','Sobrevivientes']
```

```python
df_titanic.Survived.value_counts().plot(kind = "pie", labels = survived_labels, title = "Distribución de supervivientes")
plt.show()
```

```python
plt.pie(survived,labels=survived_labels,explode=(0, 0.1),autopct='%1.1f%%',shadow=True)
plt.title("Distribución de supervivientes")
plt.show()
```

## 2. Histograma con las edades.


```python
df_titanic.Age.plot(kind = "hist",bins=8,title = "Histograma de edades",color='lightblue',rwidth=0.95,)
plt.show()
```

```python
df_titanic.Age.describe()
```

```python
plt.hist(x=df_titanic['Age'],bins=80,color='purple', rwidth=0.95,)
plt.title('Histograma de edades')
plt.show()
```

## 3. Diagrama de barras con el número de personas en cada clase.


```python
df_titanic.Pclass.value_counts().plot(kind = "bar", title = "Número de personas por clase")
plt.show()
```

```python
df_titanic.groupby("Pclass").size().plot(kind = "bar", title = "Número de personas por clase")
plt.show()
```

```python
df_titanic.groupby('Pclass')['Age'].count().to_list()
```

```python
df_titanic.groupby('Pclass')['Age'].value_counts()
```

## 4. Diagrama de barras con el número de personas fallecidas y supervivientes en cada clase.

```python
df_titanic.groupby(["Pclass", "Survived"]).size().unstack()
```

```python
df_titanic.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar")
plt.title("Número de personas fallecidas y supervivientes por clase")
plt.show()
```

## 5. Diagrama de barras con el número de personas fallecidas y supervivientes acumuladas en cada clase.

```python
df_titanic.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True)
plt.title("Número de personas fallecidas y supervivientes por clase")
plt.show()
```
