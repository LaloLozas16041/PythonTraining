# El cofre de Isla Lejana


Fardelejos.png

Far, Dele y Jitos volvieron al cabo de una hora. Al verlos venir desde lejos, Pyratilla y su tripulación decidieron esperarles en tierra.

Cuando la guardia real llegó a la altura de los Pyrates, el primero en hablar fue Dele: "Nos complace anunciaros que nuestro rey ha dado luz verde a vuestra estancia en Isla Lejana."

"De hecho, se vio muy conmovido por vuestras palabras", añadió Jitos. Pyratilla, Pym y Pyerce se alegraron de oir la noticia. Estaban ansiosos de empezar su exploración cuanto antes.

No obstante, la guardia real los retuvo un poco más. Esta vez, quien habló fue Far diciendo: "Como ya os comunicamos a vuestra llegada, tenéis un total de 3 días. Pasado ese tiempo, si aún seguís en la isla, os tendremos que detener por haber inclumplido la ley."

"Muchas gracias por el recordatorio, Far. Lo tendremos en cuenta. ¿Hay alguna otra ley de la que tengamos que ser conscientes para evitar incumplirla y acabar en los calabozos del castillo?", preguntó Pyratilla.

Dele fue quien contestó: "Aparte de no cometer ningún crimen (robar, herir...) y el plazo de 3 días, creo que vuestra estancia será tranquila y no tendréis ningún problema con la ley."

"Eso sí, como el rey de los Fardelejos ha aceptado vuestra presencia en Isla Lejana, durante estos 3 días sois considerados habitantes de la isla, lo que implica que en caso de que alguien os robase, dañase alguna de vuestras propiedades u os infligiera algún daño físico o emocional, estáis en pleno dereho de presentar una denuncia contra el delincuente, al que se lo castigará de acuerdo a su crimen.", añadió Jitos.

"Muchas gracias, guardia real. No os causaremos ningún problema ni a vosotros ni a ningún habitante de esta hermosa isla." dijo Pyratilla. "Pym, Pyerce, preparad las cosas! Vamos de excursión por Isla Lejana."

Dicho esto, la guardia real y los Pyrates se separaron. La guardia real volvió al castillo mientras que los Pyrates se internaron en la isla, con su objetivo secreto de encontrar el cofre del tesoro oculto.
