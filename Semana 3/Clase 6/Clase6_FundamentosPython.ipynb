```python
import numpy as np
import matplotlib.pyplot as plt
```

### Funcion lineal

```python
x = np.arange(-11, 11)
y1 = x
y2=2*x+5
print(x)
print(y2)
```

```python
plt.plot(x,y1,color='blue')
plt.plot(x,y1,color='orange', linestyle='--')

plt.show()
```

```python
# Plot clasico
plt.axhline(0,linestyle=':',color='gray')
plt.axvline(0,linestyle='--', color="black")
plt.plot(x, y1)
plt.plot(x, y2)
plt.title("Titulo : Funcin Lineal")
plt.ylabel("Y")
plt.xlabel("X")
plt.xlim(-10,0)
plt.ylim(-20,10)
plt.grid(True)
plt.show()
```

```python
#Los metodos axhline y axvline generan las lineas negras en forma de cruz
#En vez de utilizar al metodo plot como el ejemplo anterior utilizamos scatter()
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.title("Titulo : Funcin Lineal")
plt.ylabel("Y")
plt.xlabel("X")
plt.xlim(-10,0)
plt.ylim(-20,10)
plt.grid(True)
plt.show()
```

### Funciones trigonometricas

```python
#plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y1 = 4 + 2 * np.sin(2 * x)
y2 = np.cos(x)

# plot
fig, ax = plt.subplots()
plt.plot(x, y1, linewidth=2.0)
plt.plot(x, y2, linewidth=1.0)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),ylim=(0, 8), yticks=np.arange(1, 8),title='funcion seno')
ax.set(xlim=(0, 5))
#ax.set(xticks=np.arange(1, 8))
#ax.set(ylim=(0, 8))
#ax.set(yticks=np.arange(1, 8))
plt.show()
```

```python
# make the data
np.random.seed(500)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 90, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
