### Ejercicio 1
Escribir dos funciones que simulan una calculadora científica que permita calcular el seno, coseno, tangente, exponencial y logaritmo neperiano o natural (logaritmo en base euler). 


## A)
* La función uno preguntará al usuario el intervalo donde quiere aplicar la función matemática y la función a aplicar, y mostrará por pantalla una tabla donde la primera columna será los valores del intervalo dado (x) y la columna 2 tendrá los resultados de el resultado de aplicar la función a este intervalo (y).


```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def funcion():
    f=0
    while f!= 1 and f!= 2 and f!= 3 and f!= 4 and f!= 5 : 
        f= int(input('Que funcion deseas aplicar?  \n1: Seno\n2: Coseno\n3: Tangente\n4:Exponencial\n5:Log natural '))
        
    rang=int(input('Digita el limite inferior del rango que deseas aplicar esa funcion: '))
    rang1=int(input('Digita el limite superior del rango que deseas aplicar esa funcion: '))
    x=np.arange(rang,rang1+1)#Debo considerar que la persona tambien quiere ese ultimo digito
    y=x 
    
    if f==1:
        y2=np.sin(x)
    elif f==2:
        y2=np.cos(x)
    elif f==3:
        y2=np.tan(x)
    elif f==4:
        y2=np.exp(x)
    else:
        y2=np.log(x) 
    data={ 'X': x, 'Y': y2}
    print(pd.DataFrame(data))
    return (x,y2,f)

x,y2,f=funcion()
```

### Explicacion 
Primero tenia que validar que el usuario digite una funcion especifica, para ello meti un bucle para que digite alguna funcion de manera correcta, de modo contrario pues se repetira el mensaje de que introduzca la funcion. 

Una vez hecho eso, con Numpy cree dos arrays, quienes almacenaran los datos de la tabla para posteriormente convertirlo en un dataframe, se programo un caso para cada funcion y se dejo la creacion del data frame para lo mero ultimo, asi ahorro mucho codigo creando un dataframe para cada caso. 

## B)
* La función dos graficara con matplotlib el resultado de función anterior y la mostrará en pantalla. Recuerden poner los títulos ejes y demás configuraciones de las gráficas para que queden de la mejor manera posible.


```python
def graficas(x,y2,f):
    plt.plot(x, y2)
    if f==1:
        plt.title("Titulo : Función Seno")
    if f==2:
        plt.title("Titulo : Función Coseno")
    if f==3:
        plt.title("Titulo : Función Tangente")
    if f==4:
        plt.title("Titulo : Función exponencial")
    if f==5: 
        plt.title("Titulo : Función Logaritmo natural")
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.show()
    return 
graficas(x,y2,f)
```

### Explicacion
Esta parte fue mucho mas facil, ya que solo era mandarle a la funcion, los parametros que habiamos obtenido anteriormente, una vez hecho eso graficamos el plot, para las etiquetas pues recordemos que habian 5 casos, asi que hacemos varios if para cada uno de los casos y de esta manera tenemos el plot correcto con su leyenda correcta.

# Ejercicio 2
Escribir un programa para ver los beneficios de una empresa en un periodo de años.
El programa debe cumplir los siguientes requisitos:
## A)
* El programa tiene que preguntar al usuario por un año inicial y otro final, y después preguntar por los ingresos y egresos de cada año desde el año inicial hasta el año final.
* Debe garantizar que por lo menos haya 5 años entre el año inicial y el final.


```python
ran1 = int(input('Ingresa el año inicial: '))
ran2 = int(input('Ingresa el año final: '))
dif = ran2 - ran1

while dif<4:
    ran1 = int(input('Ingresa el año inicial: '))
    ran2 = int(input('Ingresa el año final: '))
    dif = ran2 - ran1
rango =np.arange(ran1,ran2+1)
ing = []
eg = []
for i in rango:
    ingresos=float(input(f'Digita el ingreso en el año {i}:'))
    ing.append(ingresos)
    egresos=float(input(f'Digita el egreso en el año {i}:'))
    eg.append(egresos)
    
```

### Explicacion
Primero solicitamos los años cuales definiran mi rango, despues calculamos la diferencia y nos aseguramos de que esa diferencia sea alm menos de 5. 
Una vez hecho, para cada year solicitaremos los ingresos y los egresos mediante un ciclo for en el rango propuesto.

## B)
* Con los datos introducidos se deben crear un dataframe, con las columnas fecha, ingresos y egresos.


```python
data= {'Fecha': rango, 'Ingresos': ing, 'Egresos': eg}
df=pd.DataFrame(data)
df #Nota, en todos los years los egresos son 1
```

### Explicacion
Creamos el data set con el rango , ingresos y egresos mediante un diccionario y la libreria Pandas

## C)

```python
df['ganancias'] = df['Ingresos'] - df['Egresos']
df
```

### Explicacion
Se crea una nueva columna en el dataset, el cual va a restarle los egresos a los ingresos, de esta manera podemos calcular las ganancias, todo esto se facilita demasiado con la libreria pandas.

## D) 
* El programa debe crear otra columna llamada ganancia la cuál indique para cada año si ha habido beneficios o no de manera booleana y mostrarla por pantalla.

```python
df['Ganancia']= df['ganancias']> 0
df
```
