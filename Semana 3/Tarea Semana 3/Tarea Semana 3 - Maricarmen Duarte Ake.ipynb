Ejercicio 1: Escribir dos funciones que simulan una calculadora científica que permita calcular el seno, coseno, tangente, exponencial y logaritmo neperiano o natural (logaritmo en base euler). 

a) La función uno preguntará al usuario el intervalo donde quiere aplicar la función matemática y la función a aplicar, y mostrará por pantalla una tabla donde la primera columna será los valores del intervalo dado (x) y la columna 2 tendrá los resultados de el resultado de aplicar la función a este intervalo (y).

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

def trigo_calc(interval0, interval1, function):
    try:
        interval0 = float(interval0)
        interval1 = float(interval1)
        if interval0 < 0 or interval0 > 360 or interval1 < 0 or interval1 > 360:
            raise ValueError("El valor debe estar entre 0 y 360")
    except ValueError:
        print("El valor debe estar entre 0 y 360")
        return None

    if function not in ["1", "2", "3"]:
        print("Función inválida, intente de nuevo")
        return None

    interval_radians = np.deg2rad(np.arange(interval0, interval1 + 5, 5))

    if function == "1":
        result = np.sin(interval_radians)
    elif function == "2":
        result = np.cos(interval_radians)
    elif function == "3":
        result = np.tan(interval_radians)

    return interval_radians, result

interval0 = input("¿En qué intervalo inicial quiere aplicar la función? (De 0 a 360°)")
interval1 = input("¿En qué intervalo final quiere aplicar la función? (De 0 a 360°)")
function = input("Seleccione 1 para Seno, 2 para Coseno, 3 para Tangente")

interval_radians, result = trigo_calc(interval0, interval1, function)

if interval_radians is not None and result is not None:
    if function == "1":
        y_label ="Seno"
    elif function == "2":
        y_label = "Coseno"
    elif function == "3":
        y_label = "Tangente"
    df_trigo = pd.DataFrame({'Grados': interval_radians, y_label: result})
    print(df_trigo)
```

i) Gráfica el diagrama de barras del porcentaje de modelos de cada marca.

```python
import matplotlib.pylab as plt
# crosstab de frecuencias
crosstab = pd.crosstab(coches['Marca'], coches['Modelo'])

# convertir frecuencias a porcentajes
percentages = crosstab.apply(lambda r: r/r.sum(), axis=1)*100

# gráfica de barras
ax = percentages.plot(kind='bar', stacked=True, figsize=(10,6))
ax.set_ylabel('Porcentaje')
ax.set_ylim([0, 100])
ax.legend(loc='upper right')

plt.show()
```
