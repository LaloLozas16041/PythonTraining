### Ejercicio 1

Escribir dos funciones que simulan una calculadora científica que permita calcular el seno,
coseno, tangente, exponencial y logaritmo neperiano o natural (logaritmo en base euler).

A. La función uno preguntará al usuario el intervalo donde quiere aplicar la función
matemática y la función a aplicar, y mostrará por pantalla una tabla donde la primera
columna será los valores del intervalo dado (x) y la columna 2 tendrá los resultados de
el resultado de aplicar la función a este intervalo (y).

B. La función dos graficara con matplotlib el resultado de función anterior y la mostrará en
pantalla. Recuerden poner los títulos ejes y demás configuraciones de las gráficas para
que queden de la mejor manera posible.

```python
import math
import matplotlib.pyplot as plt

def intervalo():
    valor1=int(input("Ingrese un valor entero de inicio del Intervalo:"))
    valor2=int(input("Ingrese un valor entero de final del Intervalo:"))
    funcion=int(input("Tipo de funcion a aplicar: (1)seno,(2)coseno,(3)tangente,(4)exponencial o (5)logaritmo natural:"))
    resultados={}
    noexiste=0
    for i in range(valor1,valor2+1):
        if funcion==1:
            resultados[i]=round(math.sin(i),2)  
        elif funcion==2:
            resultados[i]=round(math.cos(i),2)
        elif funcion==3:
            resultados[i]=round(math.tan(i),2)
        elif funcion==4:
            resultados[i]=round(math.exp(i),2)
        elif funcion==5:
            resultados[i]=round(math.log(i),2)
        else:
            noexiste=1
            print("No existe la funcion",noexiste)        
    print("X","Y") 
    for clave, valor in resultados.items():
            print(clave,valor)
    x = list(resultados.keys())
    y = list(resultados.values())
    return x,y

def grafica(a,b):
    print(a)
    print(b)
    plt.plot(a,b,color='red', linestyle='--')
    plt.title("Titulo : GRAFICA DE FUNCION")
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.grid(True)
    plt.show()

x,y = intervalo()
grafica(x,y)
```

#### Explicación

### Ejercicio 2
Escribir un programa para ver los beneficios de una empresa en un periodo de años.
El programa debe cumplir los siguientes requisitos:

A. El programa tiene que preguntar al usuario por un año inicial y otro final, y después
preguntar por los ingresos y egresos de cada año desde el año inicial hasta el año final.
a. Debe garantizar que por lo menos haya 5 años entre el año inicial y el final.

B. Con los datos introducidos se deben crear un dataframe, con las columnas fecha,
ingresos y egresos.

C. El programa debe generar otra columna con el beneficio de cada año (ingresos menos
egresos) y mostrarla por pantalla.

D. El programa debe crear otra columna llamada ganancia la cuál indique para cada año si
ha habido beneficios o no de manera booleana y mostrarla por pantalla.

E. Finalmente el programa debe mostrar por pantalla la lista de los años con pérdidas.

```python
import pandas as pd

print("Favor de ingresar los datos del periodo al menos 5 años")
valor1=int(input("Ingrese el año de inicio:"))
valor2=int(input("Ingrese el año final:"))
fecha=[]
ingreso=[]
egreso=[]
beneficio=[]
ganancia=[]
df = pd.DataFrame()

#validar que por lo menos sean 5 años
periodo = valor2-valor1
año=0
gana = True
if periodo >= 5:
    for i in range(periodo+1):
        año = valor1+i
        fecha.append (año)       
        ingreso.append(int(input(f"Para el año,{año},Cual fue el Ingreso:")))
        ing = ingreso[i]
        egreso.append(int(input(f"Para el año,{año},Cual fue el Egreso:")))
        egr = egreso[i]
        calculo_beneficio = ing - egr
        beneficio.append(calculo_beneficio)
        
        if calculo_beneficio > 0:
            gana = True
            ganancia.append(gana)
        else:
            gana = False
            ganancia.append(gana)
            
        
else:
    print("El periodo es menor a 5 años, favor de ingresar mas años")

df['Fecha'] = fecha
df['Ingresos'] = ingreso
df['Engresos'] = egreso
df['Beneficio'] = beneficio
df['Ganancia'] = ganancia
print(df)

df[df["Ganancia"]==False]
```

```python
# A. Crear el siguiente DataFrame indexado:
import pandas as pd
dt = pd.DataFrame()
dia = ["Lunes","Martes","Miercoles","Jueves","Viernes"]
calorias = [420,380,390,490,300]
tiempo = [60,40,75,55,45]
dt["Dia"]= dia
dt["Calorias"]= calorias
dt["Tiempo"]= tiempo
dt
```

```python
# B. Calcular la media, mediana y desviación típica de ambas columnas.
dt.describe()
print("La media de calorias semanal es : ", dt['Calorias'].mean())
print("La mediana de calorias semanal es : ",dt['Calorias'].median())
print("La desviación típica de calorias semanal es : ",dt['Calorias'].std())
print()
print("La media de tiempo semanal es : ", dt['Tiempo'].mean())
print("La mediana de tiempo semanal es : ",dt['Tiempo'].median())
print("La desviación típica de tiempo semanal es : ",dt['Tiempo'].std())
dt.describe()
```

```python
# C. Añadir otra columna booleana al DataFrame para ver si se ha cumplido el reto de quemar más de 400 calorías por hora
limite = 400 / 60
dt['Reto'] = dt['Calorias']/60 > limite
dt
```

```python
#D.Filtrar el DataFrame y devolver otro DataFrame con las filas pares que cumplan que el número de calorías es mayor de 400
mascara=dt['Calorias'] > 400
new_dt=dt[mascara]
new_dt
mascara2= range(0,1,2)
new_dt.iloc[mascara2]
```

```python
# E. Añadir otra columna con los porcentajes de días que se ha conseguido el reto y los que no.
limite = 400 / 60
dt['Reto'] = dt['Calorias']/60 > limite
si=dt['Reto'].value_counts()[1]/len(dt['Reto'])*100
no=dt['Reto'].value_counts()[0]/len(dt['Reto'])*100

porcentaje=[]

for i in dt['Reto']:
#    print(i)
    if i:
        porcentaje.append(si)
    else:
        porcentaje.append(no)

dt['Porcentaje'] = porcentaje
dt
```

```python
# F. Crear un gráfico donde se muestre la progresión de las calorías y tiempo durante la semana.
import matplotlib.pyplot as plt

plt.bar(dt['Dia'],dt['Calorias'],color='red', linestyle='--')
plt.bar(dt['Dia'],dt['Tiempo'],color='blue', linestyle='-')
plt.title("Titulo : GRAFICA DE CALORIAS Y TIEMPO")
plt.ylabel("CALORIAS Y TIEMPO")
plt.xlabel("DIAS DE LA SEMANA")
plt.grid(True)
plt.show()
```
