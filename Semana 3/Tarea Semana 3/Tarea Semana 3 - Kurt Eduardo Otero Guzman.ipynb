### 1.- Escribir dos funciones que simulan una calculadora científica que permita calcular el seno, coseno, tangente, exponencial y logaritmo neperiano o natural (logaritmo en base euler). 

(A)La función uno preguntará al usuario el intervalo donde quiere aplicar la función matemática y la función a aplicar, y mostrará por pantalla una tabla donde la primera columna será los valores del intervalo dado (x) y la columna 2 tendrá los resultados de el resultado de aplicar la función a este intervalo (y).

(B)La función dos graficara con matplotlib el resultado de función anterior y la mostrará en pantalla. Recuerden poner los títulos ejes y demás configuraciones de las gráficas para que queden de la mejor manera posible.


```python
import math
import matplotlib.pyplot as plt

def funcion_uno():
    # Solicitar el intervalo y la función a aplicar
    inicio = float(input("Ingrese el inicio del intervalo: "))
    fin = float(input("Ingrese el fin del intervalo: "))
    paso = float(input("Ingrese el paso para el intervalo: "))
    funcion = input("Ingrese la función a aplicar (seno, coseno, tangente, exponencial, logaritmo): ")

    # Crear la tabla de resultados
    resultados = []
    x = inicio
    while x <= fin:
        if funcion == "seno":
            y = math.sin(x)
        elif funcion == "coseno":
            y = math.cos(x)
        elif funcion == "tangente":
            y = math.tan(x)
        elif funcion == "exponencial":
            y = math.exp(x)
        elif funcion == "logaritmo":
            y = math.log(x)
        else:
            print("Función no válida")
            return

        resultados.append((x, y))
        x += paso

    # Imprimir la tabla de resultados
    print(f"{'x':<10} {'y':<10}")
    for x, y in resultados:
        print(f"{x:<10.4f} {y:<10.4f}")

    return resultados

def funcion_dos(resultados):
    # Extraer los valores de x y y
    x_vals = [x for x, y in resultados]
    y_vals = [y for x, y in resultados]

    # Graficar los resultados
    plt.plot(x_vals, y_vals)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Gráfica de la función")
    plt.grid(True)

    # Mostrar la gráfica
    plt.show()

# Ejemplo de uso:
resultados = funcion_uno()
funcion_dos(resultados)
```

### 2.- Escribir un programa para ver los beneficios de una empresa en un periodo de años.
El programa debe cumplir los siguientes requisitos:

(A)El programa tiene que preguntar al usuario por un año inicial y otro final, y después preguntar por los ingresos y egresos de cada año desde el año inicial hasta el año final.
    
    (a)Debe garantizar que por lo menos haya 5 años entre el año inicial y el final.

(B)Con los datos introducidos se deben crear un dataframe, con las columnas fecha, ingresos y egresos.

(C)El programa debe generar otra columna con el beneficio de cada año (ingresos menos egresos) y mostrarla por pantalla.

(D)El programa debe crear otra columna llamada ganancia la cuál indique para cada año si ha habido beneficios o no de manera 
booleana y mostrarla por pantalla.

(E)Finalmente el programa debe mostrar por pantalla la lista de los años con pérdidas.


```python
import pandas as pd

# A: Solicitar año inicial y final, garantizando al menos 5 años de diferencia
anio_inicial = int(input("Ingrese el año inicial: "))
anio_final = int(input("Ingrese el año final: "))
while anio_final - anio_inicial < 5:
    print("Debe haber al menos 5 años de diferencia entre el año inicial y el final.")
    anio_inicial = int(input("Ingrese el año inicial: "))
    anio_final = int(input("Ingrese el año final: "))

# Crear una lista para almacenar los datos de cada año
datos = []

# Solicitar ingresos y egresos para cada año
for anio in range(anio_inicial, anio_final + 1):
    ingresos = float(input(f"Ingrese los ingresos para el año {anio}: "))
    egresos = float(input(f"Ingrese los egresos para el año {anio}: "))
    datos.append([anio, ingresos, egresos])

# B: Crear un dataframe con las columnas fecha, ingresos y egresos
df = pd.DataFrame(datos, columns=["fecha", "ingresos", "egresos"])

# C: Generar otra columna con el beneficio de cada año
df["beneficio"] = df["ingresos"] - df["egresos"]

# Mostrar beneficios por pantalla
print("\nBeneficios por año:")
print(df[["fecha", "beneficio"]])

# D: Crear otra columna llamada ganancia
df["ganancia"] = df["beneficio"] > 0

# Mostrar ganancias por pantalla
print("\nGanancias por año:")
print(df[["fecha", "ganancia"]])

# E: Mostrar por pantalla la lista de los años con pérdidas
print("\nAños con pérdidas:")
print(df.loc[df["beneficio"] < 0, "fecha"].tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame indexado
data = {
    "Calorias": [300, 450, 600, 350, 200, 500, 550],
    "Tiempo (min)": [50, 60, 80, 45, 30, 75, 65]
}
index = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
df = pd.DataFrame(data, index=index)

# Calcular la media, mediana y desviación típica de ambas columnas
print("Media:")
print(df.mean())
print("\nMediana:")
print(df.median())
print("\nDesviación típica:")
print(df.std())

# Añadir la columna booleana para verificar si se cumplió el reto
df["Reto cumplido"] = df["Calorias"] / df["Tiempo (min)"] * 60 > 400

# Filtrar el DataFrame y crear otro DataFrame con las filas que cumplan que el número de calorías es mayor de 400
df_filtrado = df[df["Calorias"] > 400]

# Añadir una columna con los porcentajes de días que se ha conseguido el reto y los que no
dias_reto_cumplido = df["Reto cumplido"].sum()
dias_totales = len(df)
porcentaje_cumplido = (dias_reto_cumplido / dias_totales) * 100
porcentaje_no_cumplido = 100 - porcentaje_cumplido

df_porcentajes = pd.DataFrame(
    {
        "Reto": ["Cumplido", "No cumplido"],
        "Porcentaje": [porcentaje_cumplido, porcentaje_no_cumplido]
    }
)

# Crear un gráfico que muestre la progresión de las calorías y tiempo durante la semana
fig, ax1 = plt.subplots()

ax1.set_xlabel("Día de la semana")
ax1.set_ylabel("Calorías", color="tab:red")
ax1.plot(index, df["Calorias"], color="tab:red", marker="o")
ax1.tick_params(axis="y", labelcolor="tab:red")

ax2 = ax1.twinx()
ax2.set_ylabel("Tiempo (min)", color="tab:blue")
ax2.plot(index, df["Tiempo (min)"], color="tab:blue", marker="o")
ax2.tick_params(axis="y", labelcolor="tab:blue")

fig.tight_layout()
plt.show()
```

### 5.- El fichero coches.csv contiene información sobre los modelos de coches vendidos en USA durante un determinado año. Se pide:

(A)Crear un DataFrame a partir del fichero anterior.

(B)Eliminar las filas con valores desconocidos y mostrar el número de filas del DataFrame resultante.

(C)Crear una columna con el precio en euros (cambio 1$ = 0.94€)

(D)Mostrar las 10 últimas filas del DataFrame.

(E)Mostrar el número de marcas de coches que contiene el DataFrame.

(F)Mostrar el número de modelos de cada marca que hay en el DataFrame, de mayor a menor frecuencia.

(G)Mostrar cuál es la marca y el modelo del coche más caro.

(H)Mostrar el precio medio en euros de los coches agrupando por marca y ordenando de menor a mayor precio.

(I)Gráfica el diagrama de barras del porcentaje de modelos de cada marca.

(J)Gráfica el diagrama de dispersión de la potencia y el precio.


```python
import pandas as pd
import matplotlib.pyplot as plt

# (A) Crear un DataFrame a partir del fichero "coches.csv"
df = pd.read_csv("coches.csv")

# (B) Eliminar las filas con valores desconocidos y mostrar el número de filas del DataFrame resultante
df = df.dropna()
print("Número de filas del DataFrame resultante:", df.shape[0])

# (C) Crear una columna con el precio en euros (cambio 1$ = 0.94€)
df["Precio_euros"] = df["Precio_dolares"] * 0.94

# (D) Mostrar las 10 últimas filas del DataFrame
print("\n10 últimas filas del DataFrame:")
print(df.tail(10))

# (E) Mostrar el número de marcas de coches que contiene el DataFrame
numero_marcas = df["Marca"].nunique()
print("\nNúmero de marcas de coches:", numero_marcas)

# (F) Mostrar el número de modelos de cada marca que hay en el DataFrame, de mayor a menor frecuencia
modelos_por_marca = df["Marca"].value_counts().sort_values(ascending=False)
print("\nNúmero de modelos de cada marca (de mayor a menor frecuencia):\n", modelos_por_marca)

# (G) Mostrar cuál es la marca y el modelo del coche más caro
coche_mas_caro = df.loc[df["Precio_euros"].idxmax()]
print("\nMarca y modelo del coche más caro:", coche_mas_caro["Marca"], coche_mas_caro["Modelo"])

# (H) Mostrar el precio medio en euros de los coches agrupando por marca y ordenando de menor a mayor precio
precio_medio_por_marca = df.groupby("Marca")["Precio_euros"].mean().sort_values()
print("\nPrecio medio en euros por marca (de menor a mayor precio):\n", precio_medio_por_marca)

# (I) Gráfica el diagrama de barras del porcentaje de modelos de cada marca
porcentaje_modelos_marca = (df["Marca"].value_counts(normalize=True) * 100).sort_values(ascending=False)
plt.figure(figsize=(12, 6))
porcentaje_modelos_marca.plot(kind="bar")
plt.title("Porcentaje de modelos de cada marca")
plt.xlabel("Marca")
plt.ylabel("Porcentaje")
plt.show()

# (J) Gráfica el diagrama de dispersión de la potencia y el precio
plt.figure(figsize=(12, 6))
plt.scatter(df["Potencia"], df["Precio_euros"])
plt.title("Diagrama de dispersión de la potencia y el precio")
plt.xlabel("Potencia")
plt.ylabel("Precio (Euros)")
plt.show()
```
