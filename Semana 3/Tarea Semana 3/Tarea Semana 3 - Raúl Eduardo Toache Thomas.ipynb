Ejercicio 1: Escribir dos funciones que simulan una calculadora científica que permita calcular el seno, coseno, tangente, exponencial y logaritmo neperiano o natural (logaritmo en base euler). 
La función uno preguntará al usuario el intervalo donde quiere aplicar la función matemática y la función a aplicar, y mostrará por pantalla una tabla donde la primera columna será los valores del intervalo dado (x) y la columna 2 tendrá los resultados de el resultado de aplicar la función a este intervalo (y).
La función dos graficara con matplotlib el resultado de función anterior y la mostrará en pantalla. Recuerden poner los títulos ejes y demás configuraciones de las gráficas para que queden de la mejor manera posible.


```python
import numpy as np
import matplotlib.pyplot as plt
ultimafuncion = ""

def calculadora_cientifica():
    funcion = input("Ingrese la función que desea aplicar (seno, coseno, tangente, exponencial o logaritmo natural): ")
    inicio = float(input("Ingrese el valor inicial del intervalo: "))
    ultimafuncion = funcion
    fin = float(input("Ingrese el valor final del intervalo: "))
    intervalo = np.linspace(inicio, fin, num=100)
    
    if funcion == "seno":
        resultado = np.sin(intervalo)
    elif funcion == "coseno":
        resultado = np.cos(intervalo)
    elif funcion == "tangente":
        resultado = np.tan(intervalo)
    elif funcion == "exponencial":
        resultado = np.exp(intervalo)
    elif funcion == "logaritmo natural":
        resultado = np.log(intervalo)
    else:
        print("La función ingresada no es válida.")
        return
    
    print("{:<10} {:<10}".format("x", "y"))
    for x, y in zip(intervalo, resultado):
        print("{:<10.4f} {:<10.4f}".format(x, y))
    
    return intervalo, resultado

def graficar_funcion(intervalo, resultado):
    funcion = ultimafuncion
    plt.plot(intervalo, resultado)
    plt.title("Gráfica de la función " + funcion)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

intervalo, resultado = calculadora_cientifica()
graficar_funcion(intervalo, resultado)
```

Explicación: Despues de preguntar al usuario los datos para su calculo, se usa la función linspace de numpy para crear un listado de 100 valores, todos correctamente distribuidos entre el valor inicial y el final, a los cuales se les aplicara la función seleccionada por el usuario desde el inicio. Al final de la función se imprime en pantalla "x" y "y" con una separación de 10 espacios y usando un ciclo for para imprimir los intervalos y sus resultados. Para darle buena presentación y evitar formatos amontonados use la función format especificando que para cada valor tiene un total de 10 espacios a utilizar de los cuales 4 deben ser decimales, todo alineado a la izquierda. Para que la segunda función sepa que función se calculo anteriormente se creo una variable global que toma el nombre de la función, este valor lo toma la segunda función y así le puede dar nombre a la gráfica ya los valores de intervalo y resultado ya mostraran la gráfica correspondiente.

```python
import pandas as pd

anio_inicial = int(input("Introduzca el año inicial: "))
anio_final = int(input("Introduzca el año final: "))

while anio_final - anio_inicial < 4:
    print("Debe haber al menos 5 años entre el año inicial y el año final.")
    anio_inicial = int(input("Introduzca el año inicial: "))
    anio_final = int(input("Introduzca el año final: "))

fechas = pd.date_range(start=str(anio_inicial), end=str(anio_final), freq='Y')
ingresos = []
egresos = []
for fecha in fechas:
    ingreso = float(input(f"Ingrese los ingresos del año {fecha.year}: "))
    ingresos.append(ingreso)
    egreso = float(input(f"Ingrese los egresos del año {fecha.year}: "))
    egresos.append(egreso)

datos = {'fecha': fechas, 'ingresos': ingresos, 'egresos': egresos}
df = pd.DataFrame(data=datos)

df['beneficio'] = df['ingresos'] - df['egresos']

df['ganancia'] = df['beneficio'] > 0

años_con_perdidas = df[df['beneficio'] < 0]['fecha'].dt.year.tolist()
if len(años_con_perdidas) > 0:
    print(f"Años con pérdidas: {años_con_perdidas}")
else:
    print("No hubo pérdidas en ningún año.")

print(df)
```

Explicación: Se comienza solicitando un inicio y final del intervalo de años a analizar y evaluando si cumple con el requisito de 5 años como minimo. Despues se solicita los ingresos y egresos de cada año, guardandolos dentro de listas (una para ingresos y otra para egresos). Posteriormente se crea el dataframe y se calculan y añaden al dataframe las columnas de beneficio y ganancias. Finalmente se muestran los años con perdidas y el dataframe completo.

Ejercicio 3: El fichero horas-trabajo.csv contiene el número de horas mensuales trabajadas por los empleados de una empresa durante el primer cuatrimestre. Crear un programa que realice las siguientes operaciones utilizando la librería Pandas:
Crea una columna con el número total de horas trabajadas para cada operador.
Muestra cual es el promedio de horas trabajadas en total en la empresa durante el mes Marzo
Muestra cuántos operadores hay por cada departamento de la empresa.
Muestra el número de empleados que han trabajado todos los meses, es decir, que tienen un número de horas todos los meses del cuatrimestre.


```python
import pandas as pd

df = pd.read_csv('horas-trabajo.csv', delimiter=";")
df = df.fillna(value='0')

df.iloc[:, 2:] = df.iloc[:, 2:].astype(float)
df['Total Horas'] = df.iloc[:, 2:].sum(axis=1)
df['Meses Ausente'] = (df.iloc[:, 2:6] == 0.0).sum(axis=1)

print('El promedio de horas trabajadas en Marzo fue:', df["Marzo"].mean())

operadores_por_departamento = df.groupby('Departamento')['Id'].count()
print('Operadores por departamento:')
print(operadores_por_departamento)

empleados_con_horas_todos_los_meses = (df['Meses Ausente'] == 0).sum()
print('El número de empleados que han trabajado todos los meses es:', empleados_con_horas_todos_los_meses)
```

Explicación: Importamos el archivo csv y eliminamos los espacios en blanco usando fillna(), luego con iloc se determina la conversion de string a float de las columnas 2 a la 5 que es donde estan los valores de horas trabajadas de cada mes de cada operador para finalmente poder crear la columna de "Total Horas" donde se suman los datos que convertimos a float. Decidi crear una columna llamada "Meses Ausente" para hacer más facil el proceso de distinguir que empleados estuvieron ausentes por un mes o más. Se calculo el promedio de horas trabajadas por cada operador y se agruparon por departamento para ver el tamaño de cada departamento. Al final sumamos los operadores que tuvieron cero meses ausentes y los mostramos en pantalla.

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'Día': ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes"],
    'Calorias': [420, 380, 390, 490, 300],
    'Tiempo': [60, 40, 75, 55, 45]
}

df = pd.DataFrame(datos)

calorias_mean = df['Calorias'].mean()
calorias_median = df['Calorias'].median()
calorias_std = df['Calorias'].std()
tiempo_mean = df['Tiempo'].mean()
tiempo_median = df['Tiempo'].median()
tiempo_std = df['Tiempo'].std()

print('La media de las calorias quemadas fue:', calorias_mean)
print('La mediana de las calorias quemadas fue:', calorias_median)
print('La desviación estandar de las calorias quemadas fue:', calorias_std)
print('La media de tiempo fue:', tiempo_mean)
print('La mediana de tiempo fue:', tiempo_median)
print('La desviación estandar de tiempo fue:', tiempo_std)

df['Cumplido'] = df['Calorias'] / (df['Tiempo']/60) > 400

df_filtrado = df[(df.index % 2 == 0) & (df['Calorias'] > 400)]

df['Porcentaje'] = (df['Cumplido'] == True).mean()

plt.plot(df['Día'], df['Calorias'], label='Calorias')
plt.plot(df['Día'], df['Tiempo'], label='Tiempo')
plt.legend()
plt.show()

print(df)
print(df_filtrado)
```

Ejercicio 5: El fichero coches.csv contiene información sobre los modelos de coches vendidos en USA durante un determinado año. Se pide:

Crear un DataFrame a partir del fichero anterior.
Eliminar las filas con valores desconocidos y mostrar el número de filas del DataFrame resultante.

Gráfica el diagrama de dispersión de la potencia y el precio.


```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
df = pd.read_csv('coches.csv', delimiter=";")

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
print("Número de filas: ", df.shape[0])

df['Precio_Euros'] = df['Precio'] * 0.94

print(df.tail(10))

num_marcas = df['Marca'].nunique()
print("Número de marcas de coches: ", num_marcas)

modelos_por_marca = df.groupby('Marca')['Modelo'].nunique().reset_index(name='Num_Modelos')
modelos_por_marca = modelos_por_marca.sort_values(by='Num_Modelos', ascending=False)
print(modelos_por_marca)

fila_max_precio = df.loc[df['Precio'].idxmax()]
marca_coche_mas_caro = fila_max_precio['Marca']
modelo_coche_mas_caro = fila_max_precio['Modelo']
print(f"El coche más caro es un {marca_coche_mas_caro} {modelo_coche_mas_caro}")

precio_medio_por_marca = df.groupby('Marca')['Precio_Euros'].mean().sort_values()
print(precio_medio_por_marca)

porcentaje_modelos = 100 * (modelos_por_marca['Num_Modelos'] / modelos_por_marca['Num_Modelos'].sum())
modelos_por_marca['Porcentaje_Modelos'] = modelos_por_marca['Num_Modelos'] / modelos_por_marca['Num_Modelos'].sum() * 100

plt.figure(figsize=(10, 5))
plt.bar(modelos_por_marca['Marca'], modelos_por_marca['Porcentaje_Modelos'])
plt.xticks(rotation=45, ha='right')
plt.ylabel('Porcentaje de modelos')
plt.xlabel('Marcas')
plt.title('Porcentaje de modelos de cada marca')
plt.show()

df.plot.scatter(x='Potencia', y='Precio_Euros', alpha=0.5)
plt.title('Diagrama de dispersión Potencia vs Precio')
plt.xlabel('Potencia')
plt.ylabel('Precio (Euros)')
plt.show()
```
