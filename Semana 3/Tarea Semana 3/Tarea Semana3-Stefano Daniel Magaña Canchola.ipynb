```python
#Ejercicio 1
#Escribir dos funciones que simulan una calculadora científica que permita calcular el seno, coseno, tangente, exponencial 
#y logaritmo neperiano o natural (logaritmo en base euler). 
#La función uno preguntará al usuario el intervalo donde quiere aplicar la función matemática y la función a aplicar, 
#y mostrará por pantalla una tabla donde la primera columna será los valores del intervalo dado (x) y la columna 2 tendrá 
#los resultados de el resultado de aplicar la función a este intervalo (y).

```

```python
def calculadora ():
    import math
    import pandas as pd
    import numpy
    print("Dijite el valor -a- del intervalo")
    a=int(input())
    print("Dijite el valor -b- del intervalo")
    b=int(input())
    valores_x=[]
    valores_y=[]
    fun=""
    print("¿Que funcion quiere aplicar\n"
          "1)Seno\n"
          "2)Coseno\n"
          "3)Tangente\n"
          "4)Exponencial\n"
          "5)Logaritmo Natural\n")
    funcion=int(input())
    for x in range(a,(b+1)):
        valores_x.append(x)
    if funcion==1:
        for x in range(a,(b+1)):
            x=math.radians(x)
            y=math.sin(x)
            valores_y.append(y)
            fun="-Seno-"
    elif funcion==2:
        for x in range(a,(b+1)):
            x=math.radians(x)
            y=math.cos(x)
            valores_y.append(y)
            fun="-Coseno-"
    elif funcion==3:
        for x in range(a,(b+1)):
            x=math.radians(x)
            y=math.tan(x)
            valores_y.append(y)
            fun="-Tangente-"
    elif funcion==4:
        for x in range(a,(b+1)):
            y=math.exp(x)
            valores_y.append(y)
            fun="-Exponencial-"
    elif funcion==5:
        for x in range(a,(b+1)):
            y=math.log(x)
            valores_y.append(y)
            fun="-Logaritmo Natural-"
           
    
            

    valores={"Valores X":valores_x,"Valores Y":valores_y}
    df = pd.DataFrame(valores)  
    print("La funcion seleccionada es: ",fun)
        
    
    return  df

file=calculadora()
print (file)
```

```python
import pandas as pd
import numpy as np
import matplotlib
print("Bienvenido al programa de gestion finaciera, Ingrese el año de inicio y el año final a analizar.\n")
print("Ingrese el año de inicio:")
año_inicio=int(input())
("Ingrese el año final:")
año_final=int(input())
fecha=[]
ingresos=[]
egresos=[]
beneficio=[]
ganancia=[]
perdidas=[]
if año_final-año_inicio>4:
    for n in range (año_inicio,año_final+1):
        fecha.append(n)
        print("Ingrese los ingresos del año: ",n)
        ing=int(input())
        ingresos.append(ing)
        print("Ingrese los egresos del año: ",n)
        eg=int(input())
        egresos.append(eg)
    cont=0
    for ingr in ingresos:
        cont2=0
        for egr in egresos:
            if cont==cont2:
                ben=ingr-egr
                beneficio.append(ben)
            cont2=cont2+1
        cont=cont+1
    for gan in beneficio:
        if gan>0:
            ganancia.append(True)
        elif gan<1:
            ganancia.append(False)
    cont3=0
    for perd in fecha:
        cont4=0
        for g in ganancia:
            if cont3==cont4:
                if g==False:
                    perdidas.append(perd)
            cont4=cont4+1
        cont3=cont3+1
        
    valores={"Fecha":fecha,"Ingresos":ingresos,"Egresos":egresos,"Beneficio":beneficio,"Ganancia":ganancia}
    df = pd.DataFrame(valores)  
    print("Los años con perdias son: \n",perdidas)
    print("Esta es las informacion completa: \n",df)
else:
    print("Debe de haber por lo menos 5 años de diferiencia entre el año de inicio y final.")
        
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
datos={"Dia":["Lunes","Martes","Miercoles","Jueves","Viernes"],"Calorias":[420,380,390,490,300],"Tiempo":[60,40,75,55,45]}
df=pd.DataFrame(datos)
print("El DataFrame original es:\n",df)

#Calcular la media, mediana y desviación típica de ambas columnas.
mask_cal=df.loc[:,"Calorias"]
media_cal=mask_cal.mean()
print('\nLa media de los valores de Calorias es:\n',media_cal)
mediana_cal=mask_cal.median()
print('La mediana de los valores de Calorias es:\n',mediana_cal)
desviacion_cal=mask_cal.std()
print('La desviacion tipica de los valores de Calorias es:\n',desviacion_cal)

mask_tiem=df.loc[:,"Tiempo"]
media_tiem=mask_tiem.mean()
print('\nLa media de los valores de Tiempo es:\n',media_tiem)
mediana_tiem=mask_tiem.median()
print('La mediana de los valores de Tiempo es:\n',mediana_tiem)
desviacion_tiem=mask_tiem.std()
print('La desviacion tipica de los valores de Tiempo es:\n',desviacion_tiem)

#Añadir otra columna booleana al DataFrame para ver si se ha cumplido el reto de quemar más de 400 calorías por hora.
#La nueva columna debe generarse aplicando una fórmula a las otras columnas. 
cont=0
meta=400/60
reto=[]
cal_min=[]
for minu in mask_tiem:
    cont1=0
    for cal in mask_cal:
        if cont==cont1:
            calxmin=cal/minu
            cal_min.append(calxmin)
            if calxmin >= meta:
                reto.append(True)
            elif calxmin < meta:
                reto.append(False)
        cont1=cont1+1
    cont=cont+1   
df["Meta cumplida"]=reto
print("\nSe agrega columna Meta cumplida al DataFrame\n",df) 

#Filtrar el DataFrame y devolver otro DataFrame con las filas que cumplan que el número de calorías es mayor de 400.
maskcal=df["Calorias"]>400
newdf=df[maskcal]
print("\nLos datos con mas de 400 cal son:\n",newdf)

#Añadir otra columna con los porcentajes de días que se ha conseguido el reto y los que no.
porc_reto=[]
for calmin in cal_min:
    porc=calmin*100/meta
    porc_reto.append(porc)
df["Porcentaje Reto %"]=porc_reto
print("\nSe agrega columna Porcentaje Reto % al DataFrame\n",df)

#Crear un gráfico donde se  muestre la progresión de las calorías y tiempo durante la semana.
time=df.loc[:,"Tiempo"]
calories=df.loc[:,"Calorias"]
date=df.loc[:,"Dia"]
days=[]
for d in date:
    days.append(d+"\nCal")
    days.append("\n"+d+"\nTime")
cal_time=[]
cont=0
for t in time:
    cont1=0
    for c in calories:
        if cont==cont1:
            cal_time.append(c)
            cal_time.append(t)
        cont1=cont1+1
    cont=cont+1

plt.bar(days,cal_time, color='blue')
plt.title('Calorias y tiempo por dia')
plt.xlabel('')
plt.ylabel('Cantidad')
plt.show()

plt.bar(date,calories, color='red')
plt.title('Calorias por dia')
plt.xlabel('')
plt.ylabel('Calorias')
plt.show()

plt.bar(date,time, color='yellow')
plt.title('Tiempo por dia')
plt.xlabel('')
plt.ylabel('Tiempo')
plt.show()


```

```python
#Crear un DataFrame a partir del fichero anterior.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
folder2=r'\Users\karen\OneDrive\Escritorio\Data Analytic Training\Python Fundamentals'
file2=r"\coches.csv"
adfile2=folder2+file2
coches=pd.read_csv(adfile2, index_col=0,delimiter=";")
print(coches)

#Eliminar las filas con valores desconocidos y mostrar el número de filas del DataFrame resultante.
filas_original=len(coches)
print("\nEl numero de filas original:\n",filas_original)
filas_drop=len(coches.dropna())
print("\nEl numero de al eliminar las vacias:\n",filas_drop)

#Crear una columna con el precio en euros (cambio 1$ = 0.94€)
precio=coches["Precio"]
euros=[]
for p in precio:
    eur=p*0.94
    euros.append(eur)
coches["Precio Euros"]=euros
print("\nSe agrega columna Precio Euros al DataFrame\n",coches)

#Mostrar las 10 últimas filas del DataFrame.
last10=coches.tail(10)
print("\nLas ultimas 10 filas son:\n",last10)

#Mostrar el número de marcas de coches que contiene el DataFrame.
list_marcas=coches.index
marcas=[]
for marca in list_marcas:
    if marca not in marcas:
        marcas.append(marca)
Num_marcas=len(marcas)
print("\nEl numero de marcas en el DataFrame es:\n",Num_marcas)

#Mostrar el número de modelos de cada marca que hay en el DataFrame, de mayor a menor frecuencia.  
modelos=[]
for mar in marcas:
    mod=coches.loc[[mar]]
    num=len(mod)
    num_mod=[mar,"=",num]
    modelos.append(num_mod)
modelos_desc=sorted(modelos, key=lambda modelos: modelos[2], reverse=True)
print("\nEl numero de modelos por marca, en orden descendente son:\n",modelos_desc)
      
#Mostrar cuál es la marca y el modelo del coche más caro.
rows=len(coches)
coche_max=coches[coches["Precio"]==coches["Precio"].max()]
print("\nEl modelo mas caro es:\n",coche_max)

#Gráfica el diagrama de barras del porcentaje de modelos de cada marca.
mar1=[]
porc1=[]
cont_mar=0
for mod_marc in modelos_desc:
    marc=mod_marc[0]
    mar1.append(marc)
    por=(mod_marc[2])*100/rows
    porc1.append(por)
plt.bar(mar1,porc1, color='red')
plt.xticks(rotation=90)
plt.title("'Porcentaje de modelos por marca\nDe un total de 100%")
plt.xlabel('Marca')
plt.ylabel('Porcentaje')
plt.show()

#Gráfica el diagrama de dispersión de la potencia y el precio.    
potencia=[]
precio=[]
for pot in range (rows):
    hp=coches.iloc[pot,2]
    potencia.append(hp)
    dollar=coches.iloc[pot,4]
    precio.append(dollar)

plt.scatter(potencia, precio, color='blue')
plt.title('Gráfico de dispersión Potencia vs Precio')
plt.xlabel('Potencia')
plt.ylabel('Precio')
plt.show()
```
